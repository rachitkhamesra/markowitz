```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('nifty50.csv')
df
```

```python
tickers = []
for col in df.columns:
    tickers.append(col)
        
tickers.remove('Date')
tickers.remove('HDFCLIFE.NS')
tickers.remove('SBILIFE.NS')
tickers
```

```python
def cagr(ticker, end, period):
    
    start = str(pd.to_datetime(end) - pd.DateOffset(years=period)).split()[0]
    if (pd.to_datetime(start)<pd.to_datetime('2016-01-01')):
        print("Recheck Dates")
        return
    end_set =0 
    approx_set=0
    
    if ticker in df:
        for i in range(len(df)-1,-1,-1):
            
            if df.loc[i, 'Date'] == end:
                #print("End Price: ", df.loc[i, ticker])
                end_price = int(df.loc[i, ticker])
                end_idx = i 
                end_set = 1 #if this becomes true, i check for last year's price
                
            if df.loc[i, 'Date'] == start:
                if (end_set==0):
                    print("Input start date invalid.")
                else:
                    start_price = int(df.loc[i, ticker])
                    #print("Start Price: ", df.loc[i, ticker])
                break
            
            elif end_set==1 and approx_set==0:
                if(end_idx-(248*period)>=0):
                    #print("Aprroximated start Price: ", df.loc[end_idx-(248*period), ticker])
                    #print("Aprroximated start date: ", df.loc[end_idx-(248*period), 'Date'])
                    start_price = int(df.loc[end_idx-(248*period), ticker])
                    approx_set=1
                elif(end_idx-247>=0):
                    #print("Aprroximated start Price: ", df.loc[end_idx-(247*period), ticker])
                    #print("Aprroximated start date: ", df.loc[end_idx-(247*period), 'Date'])
                    start_price = int(df.loc[end_idx-(247*period), ticker])
                    approx_set=1
                else:
                    #print("Aprroximated start Price: ", df.loc[end_idx-(246*period), ticker])
                    #print("Aprroximated start date: ", df.loc[end_idx-(246*period), 'Date'])
                    start_price = int(df.loc[end_idx-(246*period), ticker])
                    approx_set=1
                
        cagr = (end_price/start_price)**(1/period)-1
        return (cagr*100)
        
    else:
        return 0
```

```python
def volatility(ticker):
    arr = []
    for i in range(1,len(df)):
        arr.append(df.loc[i-1, ticker]/df.loc[i, ticker])
    
    arr = np.array(arr)
    return (arr.std()*100*math.sqrt(252))
```

```python
def sharpeRatio(cagr, volatility):
    
    val = (cagr-6)/volatility
    return val
```

```python
clm = ["Ticker", "Volatility", "CAGR", "Sharpe Ratio"]

df_new = pd.DataFrame(columns = clm)

for ticker in tickers:
    df_new = df_new.append(
        pd.Series(
            [   ticker,
             volatility(ticker),
             cagr(ticker, '2020-12-31',4),
             0             
            ],
        index = clm),
        ignore_index = True)
    
for i in range(len(df_new)):
    df_new.loc[i, "Sharpe Ratio"] = sharpeRatio(df_new.loc[i, "CAGR"], df_new.loc[i, "Volatility"])
    
df_new.sort_values("Sharpe Ratio", inplace = True, ascending = False)
df_new.reset_index(inplace = True)
df_new[:15]
```
